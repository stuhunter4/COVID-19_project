```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read COVID cases csv file into a dataframe
cases_df = pd.read_csv("Resources/statewide_cases.csv")
#put los angeles and san francisco counties into their own respective dataframes
LA_cases = cases_df.loc[(cases_df["county"] == "Los Angeles")]
SF_cases = cases_df.loc[(cases_df["county"] == "San Francisco")]

#read in county size csv for the names of counties
county_df = pd.read_csv("Resources/County_size.csv")
```

```python
#put cases and deaths columns into lists to plot for outliers
LA_case_list = LA_cases["newcountconfirmed"].tolist()
SF_case_list = SF_cases["newcountconfirmed"].tolist()

LA_death_list = LA_cases["newcountdeaths"].tolist()
SF_death_list = SF_cases["newcountdeaths"].tolist()

#generate a box plot for LA county's cases
data = [LA_case_list, LA_death_list]

red_marker = dict(marker="o", markerfacecolor="red", markersize=9)
fig, ax = plt.subplots()

ax.boxplot(LA_case_list, flierprops=red_marker)
ax.set_ylabel("Daily COVID-19 Cases")
ax.set_title("Los Angeles County")

plt.show()
```

```python
#generate box plots for each county and their respective cases and deaths
red_marker = dict(marker="o", markerfacecolor="red", markersize=9)
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4)

ax1.boxplot(LA_case_list, flierprops=red_marker)
ax1.set_ylabel("Cases")
ax1.set_title("Los Angeles County")

ax2.boxplot(SF_case_list, flierprops=red_marker)
ax2.set_ylabel("Cases")
ax2.set_title("San Francisco County")

ax3.boxplot(LA_death_list, flierprops=red_marker)
ax3.set_ylabel("Deaths")
ax3.set_title("Los Angeles County")

ax4.boxplot(SF_death_list, flierprops=red_marker)
ax4.set_ylabel("Deaths")
ax4.set_title("San Francisco County")

#tweak spacing between subplots to prevent labels from overlapping
plt.subplots_adjust(top=4, hspace=0.15)
plt.show()
```

While there are outliers since "the extreme value is a legitimate observation that is a natural part of the population you’re studying, you should leave it in the dataset."
