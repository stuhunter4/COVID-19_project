```python
#dependency
import pandas as pd
```

```python
#convert text file from https://www.census.gov/geographies/reference-files/time-series/geo/gazetteer-files.html to dataframe
#first I used " " as the delimiter but it wouldn't work, so I copy-pasted the empty space in the txt file to get the right one
data = pd.read_csv("Resources/2019_gaz_counties_06.txt", sep="	")
data
```

```python
#select only information from LA and SF counties, including land and water area only
LASF_data = data.loc[(data["NAME"] == "Los Angeles County") | (data["NAME"] == "San Francisco County")]
LASF_data = LASF_data[["NAME", "ALAND_SQMI", "AWATER_SQMI"]]
#new column with both types of area added to each other, giving total area
LASF_data["Total Area"] = LASF_data["ALAND_SQMI"] + LASF_data["AWATER_SQMI"]
#export to csv and display new data with total area
LASF_data.to_csv("Resources/County_size.csv", index=False, header=True)
LASF_data
```
