```python
#dependencies
import requests
import json
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#construct url that has the filter parameter for state_id = CA, and limit parameter to ensure 100 results are returned (the max)
url = "https://healthdata.gov/api/action/datastore/search.json?resource_id=8049ae6e-3e72-4fb4-a3b6-0fc8c85b6517&filters[state_id]=CA&limit=100"
#request the data and convert to json
data = requests.get(url).json()
data
```

```python
#pull relevant data from the json and convert into a dataframe
cali_df = pd.DataFrame(data["result"]["records"])
cali_df.drop(columns=["state_id", "total_phases"])
```

```python
county_df = cali_df.loc[(cali_df["policy_level"] == "county")]
state_df = cali_df.loc[(cali_df["policy_level"] == "state")]
```
